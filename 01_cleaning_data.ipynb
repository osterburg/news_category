```python
import re

left_space_chars = {'(', '['}
right_space_chars = {'.', ',', '!', '?', ':', ';', ')', ']'}
space_chars = {'-', '+', '=', '&', '#'}

def replace_latex(text):
    stripped = re.sub("\$\$[^$$]+\$\$", ' ', text)  # $$...$$
    stripped = re.sub("\$[^$]+\$", ' ', stripped)  # $...$
    cleared_text = re.sub(r'\\begin\{(.*?)\}(.*?)\\end\{\1\}', ' ', stripped)  # \begin...\end
    return cleared_text

def replace_newlines(text):
    return text.replace('\\n', ' ') \
            .replace('\\r', ' ') \
            .replace('\n', ' ') \
            .replace('\r', ' ') \
            .replace('\\', ' ')

def norm_spaces(text):
    for char in left_space_chars:
        text = text.replace(char, ' ' + char)

    for char in right_space_chars:
        text = text.replace(char, char + ' ')

    for char in space_chars:
        text = text.replace(char, ' ')

    return re.sub('\s+', ' ', text).strip()
```

```python
from pathlib import Path

out_file = Path('data/cleaned_news_data.csv')

train_file = Path('data/train_data.csv')
test_file = Path('data/test_data.csv')
valid_file = Path('data/valid_data.csv')

columns = ['type', 'content']
chunksize = 10000
```

```python
from tqdm import tqdm

reader = pd.read_csv(filename, usecols=columns, chunksize=chunksize, 
                     encoding='utf-8', engine='python', compression='gzip')

for i, chunk in enumerate(tqdm(reader)):
    # lower case
    chunk = chunk.apply(lambda x: x.astype(str).str.lower())
    
    # replace newlines
    chunk['content'] = chunk['content'].apply(replace_newlines)
    chunk['type'] = chunk['type'].apply(replace_newlines)
    
    # replace latex
    chunk['content'] = chunk['content'].apply(replace_latex)
    chunk['type'] = chunk['type'].apply(replace_latex)
    
    # replace links
    chunk.replace(r'^https?:\/\/.*[\r\n]*', '',  regex=True, inplace=True)
    
    # norm spaces
    chunk['content'] = chunk['content'].apply(norm_spaces)
    chunk['type'] = chunk['type'].apply(norm_spaces)
    
    # norm quotes 
    chunk.replace('\'\'', '\'', regex=True, inplace=True)
    chunk.replace('""', '"', regex=True, inplace=True)
    
    chunk.dropna(inplace=True)
    
    if 'none' in chunk.type.values:
        chunk = chunk[~chunk.type.str.contains('none')]
    
    if '2018-02-10 13:43:39.521661' in chunk.type.values:
        chunk = chunk[~chunk.type.str.contains('2018-02-10 13:43:39.521661')]
        
    # rename columns
    chunk.rename(columns={'type': 'category'}, inplace=True)
    
    # reorder columns
    chunk = chunk[['content', 'category']]

    if not out_file.exists():
        chunk.to_csv(out_file, index=False)
    else:
        chunk.to_csv(out_file, index=False, mode='a', header=False)
```
